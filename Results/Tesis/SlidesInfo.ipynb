```python
events = [
    "Riding",
    "Fighting",
    "Playing",
    "Running",
    'Lying',
    "Chasing",
    "Jumping",
    "Falling",
    "Guiding",
    "Stealing",
    "Littering",
    "Tripping",
    "Pickpockering",
]
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
rute='/home/ubuntu/Database/ALL/Videos'
```

```python
storage=dict()
for i in events:
    files=os.listdir(os.path.join(rute,i))
    storage[i]=len(files)
df = pd.DataFrame.from_dict(storage, orient='index', columns=['Number of Videos'])
print(df)
```

```python
df.plot(kind='bar', figsize=(10, 6), color='#009681')
plt.title('Distribution of the videos per Event', fontsize=14, fontweight='bold')
plt.xlabel('Events').set_visible(False)
plt.ylabel('Quantity of Videos', fontsize=12, fontweight='bold')
plt.xticks(rotation=45)
plt.legend().set_visible(False)
plt.tight_layout()
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.grid()
plt.savefig('/home/ubuntu/Tesis/Results/Tesis/distribution_videos_per_event.png', bbox_inches='tight')
plt.show()
```

```python
import cv2
for i in events:
    files=os.listdir(os.path.join(rute,i))
    for j in files:
        path=os.path.join(rute,i,j)
        cap = cv2.VideoCapture(path)
        if not cap.isOpened():
            print(f"Error opening video file: {path}")
            continue
        width= int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        if width<1280 or height<720:
            print(f"Resizing video: {path}, original size: {width}x{height}")
        cap.release()
```
